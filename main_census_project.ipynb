# 7.What is the range of the middle 50% of working hours per week? 

```python

median = census_data['hours_per_week'].median()
p25 = census_data['hours_per_week'].quantile(0.25)
p75 = census_data['hours_per_week'].quantile(0.75)

#  range of middle 50%
middle_50_range = p75 - p25

print("median is" ,median)
print('Range of middle 50%:', middle_50_range)
```
